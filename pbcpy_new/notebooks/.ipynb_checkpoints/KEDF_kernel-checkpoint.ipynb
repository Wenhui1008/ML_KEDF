```python
def rho_KE_r(pot,sigma=0.0):
    rho_KE = pot.fft()
    gg = pot.grid.get_reciprocal().gg
    rho_KE=rho_KE*gg*np.exp(-gg*sigma**2/4.0)
    rho_KE_r=rho_KE.ifft()
    return rho_KE_r
```
