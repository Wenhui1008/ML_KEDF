## Get 1D cuts

```python
#x0 = Coord([0,0,0],grid,ctype='Crystal')
x0= Coord([0,0,0],cell=Si_KS_XC.cell,ctype='Crystal')
r0 = dummy_Si.ions[1].pos - dummy_Si.ions[0].pos
pot_1d_KS_H = Si_KS_H.plot.get_plotcut(x0=x0,r0=4*r0,nr=750)
pot_1d_KS_XC = Si_KS_XC.plot.get_plotcut(x0=x0,r0=4*r0,nr=750)
pot_1d_WT_H = Si_WT_H.plot.get_plotcut(x0=x0,r0=4*r0,nr=750)
pot_1d_WT_XC = Si_WT_XC.plot.get_plotcut(x0=x0,r0=4*r0,nr=750)
#pot_1d_LDA = Si_LDA_HXC.plot.get_plotcut(x0=x0,r0=4*r0,nr=750)
pot_1d_local = local.plot.get_plotcut(x0=x0,r0=4*r0,nr=750)
pot_1d_WT_0 = WT_0.plot.get_plotcut(x0=x0,r0=4*r0,nr=750)
pot_1d_LDA_0 = LDA.plot.get_plotcut(x0=x0,r0=4*r0,nr=750)
pot_1d_WGC = WGC.plot.get_plotcut(x0=x0,r0=4*r0,nr=750)
pot_1d_KS_local = KS_locl.plot.get_plotcut(x0=x0,r0=4*r0,nr=750)
pot_1d_WT_s_HXC =WT_s_HXC.plot.get_plotcut(x0=x0,r0=4*r0,nr=750)
pot_1d_WT_s_KE = WT_s_ke.plot.get_plotcut(x0=x0,r0=4*r0,nr=750)
pot_1d_WT_s_tot = WT_s_tot.plot.get_plotcut(x0=x0,r0=4*r0,nr=750)
```

```python

fig, ax = plt.subplots(1,1,figsize=(9,6))
#ax.plot((pot_1d_KS_H.values + pot_1d_KS_XC.values)/2.0,linestyle='-',color='black',linewidth=2.5,label='KS')
#ax.plot(pot_1d_WT.values,linestyle='--',color='red',linewidth=2.5,label='WT')
KS = (pot_1d_KS_H.values + pot_1d_KS_XC.values + pot_1d_KS_local.values)/2.0
WT = pot_1d_WT_H.values + pot_1d_WT_XC.values+pot_1d_local.values
#LDA =pot_1d_LDA_0.values
#WGC =pot_1d_WGC.values
#OF_ion=pot_1d_local.values
#KS_ion=pot_1d_KS_local.values/2.0
WT_KE_P=pot_1d_WT_s_KE.values
WT_Eff_P=pot_1d_WT_s_HXC.values + pot_1d_local.values
#WT_Tot_P=pot_1d_WT_s_tot.values
shift = 5.3838/27.211385#np.average(KS[350:370])

KS = (KS-shift)*-27.211385

shift = 0.169747955849165#np.average(WT[350:370])

WT_Eff_P = (WT_Eff_P-shift)*-27.211385
WT_KE_P=WT_KE_P*27.211385
WT= (WT-shift)*-27.211385
#WT_Tot=(WT_Tot_P-shift)*27.211385
#LDA =LDA*27.211385
#WGC =WGC*27.211385




ax.plot(KS,linestyle='--',color='red',linewidth=2.5,label='$\mu-V_{eff}(KS)$')#@ax.plot(WT,linestyle='-',color='Blue',linewidth=2.5,label='$\mu-V_{eff0}(WT)$')
#ax.plot(LDA,linestyle='-',color='Green',linewidth=2.5,label='LDA')
#ax.plot(WGC,linestyle='-',color='Black',linewidth=2.5,label='WGC')
ax.plot(WT_Eff_P,linestyle='-',color='Green',linewidth=2.5,label='$\mu-V_{eff}(WT)$')
ax.plot(WT,linestyle='-',color='Black',linewidth=2.5,label='$\mu-V_{eff0}(WT)$')
#ax.plot(WT_Tot,linestyle='-',color='Yellow',linewidth=2.5,label='Tot')
#ax.plot(pot_1d_KS_local.values/2.0,linestyle='-',color='Black',linewidth=2.5,label='KE')
#ax.plot(pot_1d_local.values,linestyle='-',color='Yellow',linewidth=2.5,label='Tot')
ax.text(-15,0.000,"Si",fontsize=18)
ax.text(175,0.000,"Si",fontsize=18)
ax.text(730,0.000,"Si",fontsize=18)
#ax.text(400,0.06,"CD Si",fontsize=25)

plt.xlabel('CD Si [1 1 1] direction',fontsize=15)
#plt.ylabel('Total electron density (1/bohr$^3$)',fontsize=15)
plt.ylabel(r'$\Delta v_{Hxc}$',fontsize=15)

plt.legend(fontsize=18)
plt.hlines(y=0,xmin=0,xmax=700)
plt.savefig("./pp.eps")
plt.show()
#
```
